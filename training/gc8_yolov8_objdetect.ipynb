```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")

model.train(
    data="/content/data.yaml",
    epochs=100,
    imgsz=640,
    batch=32,
    patience=25,
    optimizer="AdamW",
    lr0=0.003,
    warmup_epochs=3,
    cache=True,
    degrees=5,
    translate=0.05,
    scale=0.1,
    shear=0.0,
    perspective=0.0,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    fliplr=0.5,
    flipud=0.0,
    mosaic=1.0,
    mixup=0.1,
    copy_paste=0.0
)
```
